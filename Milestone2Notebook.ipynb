A key variable to our project is the Intervention Type. This data consists of all types of trade measures that have been implemented since 2008. Unfortunately, there are way too many categories that make it difficult to conduct our analysis and visualization. To fix this, we thought it would be easier to categorize into two main buckets we've categorized our trade measures: Import Tariff and Non-Tariff Measure. To do this, we first created a variable "mask" that returns True if the Intervention is Import Tariff and Non-Tariff otherwise. We then used our mask variable to map this onto our new variable in the data "Trade Measure Type." Using this new variable, we now only had two categories to work with, and made it much simpler to do a small visualization on how these two measures have evovled over time.
